```python
dataset = ds.dataset('C:/Users/Utilisateur/OneDrive/Documents/centrale Lille + sc po/Projet SOGA/ess-numerique-database/data/raw/StockUniteLegale_utf8.parquet', format="parquet")
```

```python
table = dataset.to_table(
    columns=[
        'siren', 'denominationUniteLegale', 'nomUniteLegale', 'prenom1UniteLegale',
        'activitePrincipaleUniteLegale', 'economieSocialeSolidaireUniteLegale',
        'categorieJuridiqueUniteLegale', 'trancheEffectifsUniteLegale'
    ],
    filter=(
        (ds.field('activitePrincipaleUniteLegale') == '94.99Z')
    )
)
```

```python
df_insee_9499Z = table.to_pandas()
```

```python
print(f"Nombre d'acteurs 94.99Z trouvés dans l'INSEE : {len(df_insee_9499Z)}")
```

Comparaison avec entreprisess


```python
df_insee_9499Z['siren'] = df_insee_9499Z['siren'].astype(str).str.strip()
liste_siren_references = [str(s).strip() for s in liste_siren_references]
```

```python
df_final = df_insee_9499Z[df_insee_9499Z['siren'].isin(liste_siren_references)].copy()
```

```python
df_final['nom_complet'] = df_final['denominationUniteLegale'].fillna(
    df_final['nomUniteLegale'].astype(str) + ' ' + df_final['prenom1UniteLegale'].fillna('')
)
```
